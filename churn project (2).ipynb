```python
import seaborn as sns 
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.isna()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Churn']=le.fit_transform(df['Churn'])
```

```python
df.head()
```

```python
x=df.drop(['Churn','customerID'],axis=1)
```

```python
x
```

```python
y=df['Churn']
```

```python
y
```

```python
categorical_features = x.select_dtypes(include=['object']).columns
print(categorical_features)
```

```python
numerical_features = x.select_dtypes(include=['int64', 'float64']).columns
print(numerical_features )
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ])
```

```python
x['TotalCharges'] = pd.to_numeric(x['TotalCharges'], errors='coerce')
x['TotalCharges'].fillna(x['TotalCharges'].median(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_train_processed = preprocessor.fit_transform(x_train)
x_test_processed = preprocessor.transform(x_test)
```

```python
x_train_processed
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train_processed, y_train)
```

```python
y_pred=model.predict(x_test_processed)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report

score=accuracy_score(y_pred,y_test)                             
score
```

```python
print(classification_report(y_pred,y_test))
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
pca=PCA(n_components=2,random_state=42)
x_train_pca=pca.fit_transform(x_train_processed)
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x_train_pca[:, 0], x_train_pca[:, 1], c=y_train,cmap='coolwarm',edgecolor='k')
plt.title('PCA of Customer Churn Data (2D View)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(handles=scatter.legend_elements()[0], labels=['Not Churned', 'Churned'])
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
svm_model=SVC(kernel='rbf',C=1,gamma='scale',random_state=42)
svm_model.fit(x_train_processed,y_train)
```

```python
accuracy=svm_model.score(x_test_processed,y_test)
```
